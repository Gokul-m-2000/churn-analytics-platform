```python
import pandas as pd
file_path='data/Online_retail.csv'
df=pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
print(df.dtypes)
```

```python
df=df[df['Quantity']>0]
df.dropna(subset=['CustomerID'],inplace=True)
print(f"shape of cleaned dataframe : {df.shape}")
```

```python
df['CustomerID']=df['CustomerID'].astype('str')
df.info()
```

```python
basket=(df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket.head()
```

```python
basket.info()
```

```python
basket.iloc[:,-5:]
```

```python
def encodeunits(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

basket_sets=basket.applymap(encodeunits)
basket_sets.iloc[:,-5:]
```

```python
basket_sets=(basket_sets>0).astype(int)
basket_sets.iloc[:,-5:]
```

```python
basket_sets.info()
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(basket_sets,min_support=0.01,use_colnames=True,low_memory=True)
frequent_itemsets.sort_values(by='support',ascending=False).head(10)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate rules with a lift of at least 1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the rules sorted by confidence and lift
rules.sort_values(by=['confidence', 'lift'], ascending=[False, False]).head(10)
```

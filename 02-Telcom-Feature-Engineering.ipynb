```python
import pandas as pd
df=pd.read_csv('data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.info()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Churn']=df['Churn'].map({'Yes':1,'No':0})
df.head()
```

```python
df['Has_securityBundle']=((df['OnlineSecurity']=='Yes')&(df['OnlineBackup']=='Yes')).astype(int)
```

```python
premium_cols=['OnlineSecurity',
    'OnlineBackup',
   'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies']
has_premium_service=(df[premium_cols]=="Yes")
df['Premium_services']=has_premium_service.sum(axis=1)
df.info()
df.head()

```

```python
df.tail()
```

```python
df_ml=df.drop('customerID',axis=1)
df_ml=pd.get_dummies(df_ml,dtype=int)
```

```python
df_ml.head()
```

```python
df.head()
```

```python
x=df_ml.drop('Churn',axis=1)
y=df_ml['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(f"Original columns {df.shape[1]}")
print(f"df_ml columns {df_ml.shape[1]}")
print(f"x_train {x_train.shape[0]}")
```

```python
df_ml.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
print("Data successfully scaled")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report

baseline_model=LogisticRegression(max_iter=1000)
baseline_model.fit(x_train_scaled,y_train)


y_pred=baseline_model.predict(x_test_scaled)

print(f"Baseline model   -    Logistic Regression")
print(f"Accuracy score  : {accuracy_score(y_test,y_pred):.4f}")
print("Classification report")
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
xgb_model=XGBClassifier(random_state=42,learning_rate=0.1,n_estimators=100)
xgb_model.fit(x_train_scaled,y_train)
y_pred_xgb=xgb_model.predict(x_test_scaled)

print(f"model  : XGboost")
print(f"Accuracy_score   : {accuracy_score(y_test,y_pred):.4f}")
print("Classification report")
print(f"{classification_report(y_test,y_pred)}")


```

```python
import matplotlib.pyplot as plt

# 1. Get feature importance from the XGBoost model
importances = xgb_model.feature_importances_
feature_names =x.columns # This uses the column names before scaling

# 2. Create a DataFrame for easy plotting
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# 3. Plot the Top 15 features
plt.figure(figsize=(12, 8))
plt.barh(feature_importance_df['Feature'][:15], feature_importance_df['Importance'][:15], color='skyblue')
plt.xlabel('Importance Score')
plt.title('Which Features are Driving Churn Predictions?')
plt.gca().invert_yaxis() # Highest importance at the top
plt.show()
```

```python
# A more 'aggressive' XGBoost to find hidden patterns
xgb_tuned = XGBClassifier(
    n_estimators=200,      # More trees
    max_depth=6,           # Look deeper into features
    learning_rate=0.05,    # Take smaller, more careful steps
    scale_pos_weight=3,    # IMPORTANT: Tells the model Churners are 3x more important
    random_state=42
)

xgb_tuned.fit(x_train_scaled, y_train)
y_pred_tuned = xgb_tuned.predict(x_test_scaled)

print("--- Tuned XGBoost Results ---")
print(classification_report(y_test, y_pred_tuned))
```

```python
# 1. Get feature importance from the XGBoost model
importances = xgb_tuned.feature_importances_
feature_names =x.columns # This uses the column names before scaling

# 2. Create a DataFrame for easy plotting
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# 3. Plot the Top 15 features
plt.figure(figsize=(12, 8))
plt.barh(feature_importance_df['Feature'][:15], feature_importance_df['Importance'][:15], color='skyblue')
plt.xlabel('Importance Score')
plt.title('Which Features are Driving Churn Predictions?')
plt.gca().invert_yaxis() # Highest importance at the top
plt.show()
```

```python
import shap

# 1. Initialize the SHAP explainer
explainer = shap.TreeExplainer(xgb_tuned)
shap_values = explainer.shap_values(x_test_scaled)

# 2. Visualize the first customer in the test set
# This will show a 'Force Plot'
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0,:], x.iloc[0,:])
```
